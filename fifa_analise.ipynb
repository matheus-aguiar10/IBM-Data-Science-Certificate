```python
# Importando bibliotecas e lendo arquivos

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Lendo arquivo 

base = pd.read_csv('fifa.csv')
```

```python
base.columns
```

```python
base.drop(['Photo', 'Release Clause', 'Club Logo', 'Unnamed: 0', 'Flag'], axis = 1, inplace = True)
```

```python
base.head()
```

```python
#Pegando os 20 melhores jogadores

melhores = base.sort_values(by = "Overall",ascending = False)
melhores.head(20)
```

```python
#Editando nomes das colunas

base.columns.str.replace(' ', '_')
```

```python
#Quantidade de jogadores LEFT and RIGHT FOOT

sns.countplot(data=base, x=base['Preferred Foot'])
```

```python
#Média do RATING dos jogadores

media = base.groupby('Nationality').Overall.mean().sort_values(ascending = False)
plt.figure(figsize = (18,8))
media.head(60).plot(kind = 'bar')
```

```python
#Distribuicão de positions

plt.figure(figsize=(15,8))
sns.countplot(data=base, x=base['Position'])
```

```python
#Distribuição por nacionalidade

nacionalidades = base.Nationality.value_counts(ascending = False)
plt.figure(figsize = (14,8))
nacionalidades.head(50).plot(kind = 'bar')
```

```python
#Tratando a coluna dos valores

base['valores'] = base.Value.str.replace('€', '').str.replace('K', '').str.replace('M', '').astype(float)
```

```python
#Distribuicao por valor dos jogadores

sns.distplot(a = base.valores, bins = 5, color = 'black')
```

```python
#Tratando coluna de salários

base['salarios'] = base.Wage.str.replace('€', '').str.replace('K', '').str.replace('M', '').astype(float)
```

```python
#Salarios dos jogadores entre 100K e 500k

salarios1 = base[(base.salarios >= 100.00) & (base.salarios <= 500.00)]
```

```python
#Vendo estatisticas de quem tem salario entre 100K e 500K

salarios1.salarios.describe()
```

```python
#Distribuiçao salarios entre 100K e 500K

sns.distplot(a = salarios1.salarios, bins = 10, color = 'blue')
```

```python
base['valores'] = base.Value.str.replace('€', '').str.replace('K', '000').str.replace('M', '000000').astype(float)
```

```python
#10 jogadores mais caros

caros = base.valores.sort_values(ascending = False)
plt.figure(figsize = (14,8))
caros.head(10).plot(kind = 'bar')
```
