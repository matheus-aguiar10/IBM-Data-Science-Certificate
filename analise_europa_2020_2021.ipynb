```python
#Importando bibliotecas

from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Verificando se a requisicao na pagina foi bem sucedida
#Utilizando metodo GET para fazer a requisicao

req = requests.get('https://fbref.com/en/comps/Big5/Big-5-European-Leagues-Stats.html')
if req.status_code == 200:
    print('Requisição bem sucedida!')
    content = req.content
```

```python
#Utilizando o soup para extrair a tabela

soup = BeautifulSoup(content, 'html.parser')
table = soup.find(name='table')
```

```python
#Antes de passar a variável ‘table’ na função, devemos convertê-la para string primeiro
#O retorno deste método é sempre uma lista de Data Frames, e portanto devemos acessar a posição 0 dela para obter nossa tabela.

table_str = str(table)
df = pd.read_html(table_str)[0]
```

# Legenda de cada Coluna

Squad = Nome do Time

Country = Liga

LgRk = Posição na liga

MP = Numero de jogos

W = Numero de vitorias

D = Numero de empates

L = Numero de derrotas

GF = Gols feitos

GA = Gols sofridos

GDiff = Saldo de Gols

Pts = Pontos ganhos

Pts/G = Pontos por jogo

xG = Expectativa de gols

xGA = Expectativa de gols permitido

xGDiff = Expectativa saldo de gols

xGDiff/90 = Expectativa saldo de gols por 90 minutos

Last 5 = Últimos 5 jogos

Attendance = Público presente

Top Team Scorer = Artilheiro

Goalkeeper = Goleiro

```python
#Avaliando os 10 teams que mais fizeram gols na temporada 2020/2021

df1 = df.sort_values(by = ['GF'], axis=0, ascending = False)
df1[['Squad', 'GF','MP']].head(10)
```

```python
#Avaliando os 10 teams que tem a maior expectativa de gols

df2 = df.sort_values(by = ['xG'], axis=0, ascending = False)
df2['Media_gols_por_jogo'] = df2['GF']/df2['MP']
df2[['Squad', 'xG', 'GF', 'MP']].head(10)
```

```python
#Avaliando os 10 teams que tem a maior média de gols na temporada 2020/2021

df3 = df2.sort_values(by=['Media_gols_por_jogo'], axis=0, ascending = False)
df3[['Squad', 'Media_gols_por_jogo', 'MP']].head(10)
```
